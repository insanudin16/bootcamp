```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='wilayah_pengadilan_agama', data=df)
plt.xlabel("Kabupaten")
plt.xticks(rotation=90)
plt.title("Kasus Perceraian berdasarkan Wilayah")
plt.show()
```

## Groupby Untuk Jumlah Kasus perwilayah

```python
data_per_kab_kota = df.groupby('wilayah_pengadilan_agama') ['jumlah_perceraian'].sum()
```
